## 3. Верно ли, что студенты, имеющие больше 10 пропусков, учатся хуже, чем остальные?

```python
a = data[data.absences > 10].G3.mean()
a
```

```python
b = data[data.absences < 10].G3.mean()
b
```

```python
if a < b:
    print('true')
else:
    print('false')
```

## 4. Как вы думаете, какие еще признаки могут повлиять на итоговую оценку? Найдите закономерности, аналогичные рассмотренным.

### Как влияет желание получить высшее образование?

```python
a = data[data.higher == 'yes'].G3.mean()
a
```

```python
b = data[data.higher == 'no'].G3.mean()
b
```

```python
if a > b:
    print('Студенты желающие получить высшее образование учатся лучше')
else:
    print('Студенты не желающие получить высшее образование учатся лучше')
```

### Как влияет посещение платных кружков на среднюю оценку?

```python
a = data[data.paid == 'yes'].G3.mean()
a
```

```python
b = data[data.paid == 'no'].G3.mean()
b
```

```python
if a > b:
    print('Посещение платных занятий влияет на оценку')
else:
    print('Посещение платных занятий не влияет на оценку')
```


## 5. Как вы думаете, если при школе откроется общежитие, насколько актуально это будет для студентов математического класса?


```python
traveltime = data.traveltime >=3
rural = data.address == 'R'
if len(data[traveltime | rural]) / len(data)>0.49:
    print('Актуально')
else:
    print('Не актуально')
```

## 6. Найдите в данных закономерности с помощью группировки.

### Люди в отношениях учатся хуже 

```python
data.groupby('romantic').G3.mean()
```

### Ученики, чьи родители в разводе - учатся лучше

```python
data.groupby('Pstatus').G3.mean()
```
